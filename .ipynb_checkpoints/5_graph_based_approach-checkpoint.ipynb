# Create graph

```python
from torch_geometric.data import HeteroData
from torch_geometric.transforms import RandomLinkSplit, ToUndirected
```

### Modify datasets to fit graph requirements

Pytorch geometric uses index from 0 to x

So our ids need to start from 0 and go up to x

```python
# Adjust USER_INFO
# UserId starts from 1 and goes uniterrupted up to max(userId), so we just need to move the column one number down
ratings['userId'] = ratings['userId'].apply(lambda x: x-1)

#adjust MOVIE_INFO
# Create a dict for movieId's - we will use an index for movies
movie_index_dict = {}
for movie_id in movie_info['id']:
    if movie_id not in movie_index_dict.keys():
        movie_index_dict[movie_id] = len(movie_index_dict)
    else:
        print('Error - duplicate movie ids')
        
# Add movie_graph_id to ratings and movie_info
ratings['movie_graph_id'] = ratings['movieId'].apply(lambda x: movie_index_dict[x])
movie_info['movie_graph_id'] = movie_info['id'].apply(lambda x: movie_index_dict[x])

# Sort movies by 'movie_graph_id' (unnecesary but better safe than sorry)
movie_info.sort_values(by=['movie_graph_id'], inplace=True)
```

```python
# Remove non-numeric columns from movie_info
movie_info_reccomend = movie_info.copy() #later for recommendations (we need to have title to display to user)
movie_info = movie_info.select_dtypes(['number'])
print(f'Number of numeric columns: {len(movie_info.columns)}')
```

### Create graph


```python
user_ids = np.array(ratings['userId'])
movie_ids = np.array(ratings['movie_graph_id'])
rating_ids = np.array(ratings['rating'])
```

```python
data = HeteroData()
#Node features
data['user'].x = torch.tensor(np.array(user_info.drop(columns=['userId'])))
data['movie'].x = torch.tensor(np.array(movie_info.drop(columns=['id', 'movie_graph_id'])))

#Edges (simultainously user and movie nodes, already with their types)
data['user', 'rating', 'movie'].edge_index = torch.stack([torch.tensor(user_ids, dtype=torch.long),
                                                          torch.tensor(movie_ids, dtype=torch.long)])

#Edge weights
data['user', 'rating', 'movie'].edge_label = torch.tensor(rating_ids, dtype=torch.long)
data['user', 'rating', 'movie'].edge_attr  = torch.tensor(rating_ids, dtype=torch.long)

data = ToUndirected()(data)
# del data['movie', 'rev_rating', 'user'].edge_label
```

### Train test split

```python
# 2. Perform a link-level split into training, validation, and test edges.
torch.manual_seed(123)

transform = RandomLinkSplit(
    num_val=0.1,
    num_test=0.1,
    key='edge_label',
    is_undirected=True,
    
    neg_sampling_ratio=0.0,
    add_negative_train_samples=False,
    
    edge_types=[('user', 'rating', 'movie')],
    rev_edge_types=[('movie', 'rev_rating', 'user')],
)

train_data, val_data, test_data = transform(data)
print(train_data)
print(val_data)
print(test_data)
```

# Model

```python
from torch_geometric.nn import GATConv, Linear, to_hetero, HeteroConv, GeneralConv, BatchNorm
import torch.nn as nn
import torch.nn.functional as F

class Encoder(torch.nn.Module):
    def __init__(self, hidden_channels, out_channels):
        super().__init__()
        self.conv1 = GATConv((-1, -1), hidden_channels, edge_dim=1, heads=3)
        self.conv2 = GATConv(hidden_channels*3, hidden_channels, edge_dim=1, heads=3)
        self.conv3 = GATConv(hidden_channels*3, out_channels, edge_dim=1)

    def forward(self, x, edge_index, edge_weight):
        x = self.conv1(x, edge_index, edge_attr=edge_weight)
        x = F.relu(x)
        x = self.conv2(x, edge_index, edge_attr=edge_weight)
        x = F.relu(x)
        x= self.conv3(x, edge_index, edge_attr=edge_weight)
        return x

class Predictor(torch.nn.Module):
    def __init__(self, input_dim, hidden_dim, out_dim):
        super().__init__()
        self.fc1 = nn.Linear(input_dim*2, hidden_dim)
        self.fc2 = nn.Linear(hidden_dim, hidden_dim)
        self.fc3 = nn.Linear(hidden_dim, out_dim)

    def forward(self, x, edge_label_index):
        row, col = edge_label_index
        x = torch.cat([x['user'][row], x['movie'][col]], dim=-1)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x
      
    
class GAT(torch.nn.Module):
    def __init__(self, hidden_channels_encoder, latent_space_dim, hidden_channels_predictor, n_classes):
        super().__init__()
        self.encoder = Encoder(hidden_channels_encoder, latent_space_dim)
        self.encoder = to_hetero(self.encoder, train_data.metadata(), aggr='sum')
        self.predictor = Predictor(latent_space_dim, hidden_channels_predictor, n_classes)

    def forward(self, x, edge_index, edge_label_index, edge_weight):
        x = self.encoder(x, edge_index, edge_weight)
        x = self.predictor(x, edge_label_index)
        return x


model = GAT(hidden_channels_encoder=256, latent_space_dim=256, hidden_channels_predictor=256, n_classes=1)
model = model.double()
```

# Recommend - user interface

```python
def recommend(user_id, n_recommendations=10):
    
    top_5_movies = list(pd.merge(ratings[(ratings['userId'] == user_id) & (ratings['rating']==5)][:5],
                                 movie_info_reccomend[['movie_graph_id', 'title']],
                                 how = 'left', on=['movie_graph_id'])['title'])
    print('Example movies rated 5 by user:')
    for m in top_5_movies:
        print(m)
    
    # Create edge list to predict: user to every movie in dataset
    n_movies = data['movie'].x.shape[0]
    edges_to_pred = torch.tensor([np.full(n_movies, user_id), np.arange(n_movies)])
    
    # Get output of the model
    output = model(data.x_dict, data.edge_index_dict, edges_to_pred, data)
    
    # Get top x recommendations
    #preds = pd.DataFrame(output.detach().numpy(), columns=['model_output'])
    preds = pd.DataFrame(torch.max(output, dim=1)[0].detach().numpy(), columns=['model_output'])
    preds.sort_values(by=['model_output'], ascending=False, inplace=True)
    preds = preds.reset_index().rename(columns={'index':'movie_graph_id'})
    preds = pd.merge(preds, movie_info_reccomend[['movie_graph_id', 'title']], how='left', on=['movie_graph_id'])
    
    performance = preds.copy()
    performance['userId'] = user_id
    performance = pd.merge(performance, ratings[['userId', 'movie_graph_id', 'rating']], how='left', on=['userId', 'movie_graph_id'])
    performance = performance[performance['rating'].notna()]
    
    print('\n')
    for index, row in preds[:n_recommendations].iterrows():
        print(f'Recommendation {index+1}: {row["title"]}')
        
    return performance
        
details = recommend(user_id = 105, n_recommendations=5)
```
